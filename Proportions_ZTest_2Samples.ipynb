```python
import numpy as np
from statsmodels.stats.proportion import proportions_ztest # hypothesis testing
```

```python
# Proportions 2 sample test
# In 2021, On a sample of 100 students, 20 students found cheating in their examination
# Next year on a sample of 100 students, 28 found cheating in their examination, at 0.05 level of significance
# is there any significant difference in proportions of students who are cheating in both the samples?
# Null Hypothesis - proportions are same P1 = P2
# Alternative Hypothesis - proportions are different P1 is not equal to P2
number_of_cheaters = np.array([16,28])
total_samples_size = np.array([100,100])
(test_stat, p_value) = proportions_ztest(number_of_cheaters, total_samples_size, alternative='two-sided')
print(p_value)
if(p_value>0.05):
  print('failed to reject null hypothesis')
else:
 print('reject null hypothesis')
```

```python
# Perform one tail test - left tailed
# H0: P1=P2 (proportions are same)
# H1: P1<P2 (p1 is less than p2)
number_of_cheaters = np.array([36,28])
total_samples_size = np.array([100,100])
(test_stat, p_value) = proportions_ztest(number_of_cheaters, total_samples_size, alternative='smaller') # alternative='larger' for right tail
print(p_value)
if(p_value>0.05):
  print('failed to reject null hypothesis')
else:
 print('reject null hypothesis')
```
